## Top Performers
***

Upon sorting the given data set in descending order by streams, we can notice that the top performing track appears to have an incorrect value. In addition, we can see that there are tracks in the tail part of the data set that have far greater values than any of the top 5 shown in the sorted data set.

```python
df.sort_values(by = "streams", ascending = False).reset_index() ##Sorts data set in descending order according to their stream count
```

To fix this, the correspinding value for this track must be replaced and the numerical value in string data type. This is done using .loc() to call on the problem column and value. Then we can convert all the values in the column to float data types in order to accurately sort the data set. Note that using an int data type will result in an error possibly due to the large numeric values in the columns.

```python
df.loc[df["track_name"] == "Love Grows (Where My Rosemary Goes)", "streams"] = "0" ##Replaces problem track's streams value to a string value, "0"
```

```python
df["streams"] = df["streams"].astype(float) ##Converts entire streams column values to float data type
```

**The Top 5 Performing Tracks by stream count are the following:**

```python
##Filters data set to show track, artist, and streams then sorts it by streams:
df.loc[:,["track_name","artist(s)_name","streams"]].sort_values(by = "streams", ascending = False).reset_index(drop=True).head() 
```

To determine which artists are among the most frequent in the data set, the .value_counts() attribute can be used to get the number of instances of every artist in the set. This will tell us how many songs they have written that are included in the data set.

**The Top 5 Artists with the Most Number of Tracks in the Data set are:**

```python
song_count.head(6) ##Displays the top 5 artists with most songs. note that Bad Bunny and SZA are tied for third place
```

## Basic Descriptive Statistics
***

To get the mean, median, and mode of the streams column for the data set, the built-in attributes in Pandas for basic descriptive statistics can be used.

```python
##Calls on streams column from the data set, gets its mean, median, mode and assigns it in three different variables
streams_mean = df["streams"].mean()
streams_median = df["streams"].median()
streams_std = df["streams"].std()
```

```python
##Prints the output with the mean and median formatted in scientific notation with 4 decimal places
print("Streams Descriptive Statistics:") 
print("Mean: " + str("{:.4e}".format(streams_mean)))
print("Median: " + str("{:.4e}".format(streams_median)))
print("Standard Deviation: " + str("{:.4e}".format(streams_std)))
```

## Temporal Trends
***

Matplotlib can also be used to graph the temporal trends of the song releases over the years and months. To get the X an Y values, .value_counts() is once again used to get the number of instances of the values under the released_year column. This therefore corresponds to the number of tracks released in each year. X is then assigned as a variable for the indices, which in this case are the years, and Y is assigned as a variable for the number of instances, which in this case the number of tracks released per year.

```python
##Calls on released_year column then counts the instances of each year
X_year = df["released_year"].value_counts().sort_index().index ##gets the indices of the value_counts which are the years
Y_year = df["released_year"].value_counts().sort_values().values ##gets the values of the values of the indices which are the instances
print(X_year)
print(Y_year)
```

```python
##Number of Tracks Released over the Years:
plt.figure() ##initiates figure
plt.plot(X_year, Y_year) #plots values from X and Y variables which are the years and number of tracks respectively
##Formats graph:
plt.title("Number of Tracks Released over the Years") 
plt.xlabel("Year") 
plt.ylabel("No. of Tracks") 
```

This is repeated again but instead of years, the graph will show the number of tracks over the months.

```python
##Calls on released_year column then counts the instances of each year
X_month = df["released_month"].value_counts().sort_index().index ##gets the indices of the value_counts which are the months
Y_month = df["released_month"].value_counts().sort_values().values ##gets the values of the values of the indices which are the instances
print(X_month)
print(Y_month)
```

```python
##Number of Tracks Released over the Months:
plt.figure() ##initiates figure
plt.plot(X_month, Y_month) #plots values from X and Y variables which are the months and number of tracks respectively
##Formats graph:
plt.title("Number of Tracks Released over the Months") 
plt.xlabel("Month") 
plt.ylabel("No. of Tracks") 
plt.xticks(range(1,13)) ##added to show all months of the year
```

## Genre and Music Characteristics
***

Using a scatterplot graph through Matplotlib, the correlation between the various attributes of the tracks in the data set can be examined. Variables are first assigned for the different values of the attributes in the data set. Then a scatterplot is generated and formatted. For a better analysis, the co-efficient of regression is also computed using the .corr() attribute in Pandas. This process is then repeated to different combinations of attributes:

```python
##Declaring Variables for Columns:
streams = df["streams"]
bpm = df["bpm"]
danceability = df["danceability_%"]
energy = df["energy_%"]
valence = df["valence_%"]
acousticness = df["acousticness_%"]
```

```python
##Streams as a Function of BPM:
plt.figure() ##initiates figure
plt.scatter(bpm, streams,s=3,color="green") #plots values 
##Formats graph:
plt.title("Streams as a Function of BPM") 
plt.xlabel("BPM") 
plt.ylabel("No. of Streams (Billions)") 
##Compute co-efficient of regression:
print("Regression Co-efficient: ")
df[["bpm","streams"]].corr()
```

```python
##Streams as a Function of Danceability
plt.figure() ##initiates figure
plt.scatter(danceability, streams,s=3) #plots values 
##Formats graph:
plt.title("Streams as a Function of Danceability") 
plt.xlabel("Danceability (%)") 
plt.ylabel("No. of Streams (Billions)") 
##Compute co-efficient of regression:
print("Regression Co-efficient: ")
df[["danceability_%","streams"]].corr()
```

```python
##Streams as a Function of Energy
plt.figure() ##initiates figure
plt.scatter(energy, streams,s=3,color="purple") #plots values 
##Formats graph:
plt.title("Streams as a Function of Energy") 
plt.xlabel("Energy (%)") 
plt.ylabel("No. of Streams (Billions)") 
##Compute co-efficient of regression:
print("Regression Co-efficient: ")
df[["energy_%","streams"]].corr()
```

```python
##Danceability as a Function of Energy
plt.figure() ##initiates figure
plt.scatter(energy, danceability,s=3,color="red") #plots values 
##Formats graph:
plt.title("Danceability as a Function of Energy") 
plt.xlabel("Energy (%)") 
plt.ylabel("Danceability (%)") 
##Compute co-efficient of regression:
print("Regression Co-efficient: ")
df[["energy_%","danceability_%"]].corr()
```

```python
##Valence as a Function of Acousticness
plt.figure() ##initiates figure
plt.scatter(acousticness, valence,s=3,color="orange") #plots values 
##Formats graph:
plt.title("Valence as a Function of Acousticness") 
plt.xlabel("Acousticness (%)") 
plt.ylabel("Valence (%)") 
##Compute co-efficient of regression:
print("Regression Co-efficient: ")
df[["acousticness_%","valence_%"]].corr()
```

## Advanced Analysis
***

.pivot_table() is an attribute in Pandas that can be used to check how tracks with the same keys or modes fair in terms of their streams and inclusion in spotify playlists respectively. With this, the required columns, which in this cases are key to streams and artist to in_spotify_playlitsts, can be called and pivoted to yield the attributes that are similar and the mean of their values.

```python
##Calls on key and streams values then pivots the table to get the mean streams for every key. this is then sorted by streams
df.loc[:,["key","streams"]].pivot_table(index="key", values="streams").sort_values(by="streams",ascending=False)
```

```python
##Calls on mode and streams values then pivots the table to get the mean streams for every mode this is then sorted by streams
df.loc[:,["mode","streams"]].pivot_table(index="mode", values="streams").sort_values(by="streams",ascending=False)
```

```python
##Calls on artist name and in_spotify_playlists values for solo-artists:
solo_artists = df.loc[df["artist_count"] == 1,["artist(s)_name","in_spotify_playlists"]] 
##Pivots the table in order to get the average in_spotify_playlists values for the tracks of every solo artist in the data set:
solo_artists.pivot_table(index="artist(s)_name", values="in_spotify_playlists").sort_values(by="in_spotify_playlists",ascending=False).head(10)
```
